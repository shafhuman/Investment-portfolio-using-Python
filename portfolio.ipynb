```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('Stocks2-closep.csv')
ds.head()
```

```python
ds1 = ds.set_index('timestamp')
ds1.head()
```

```python
ds1.plot(grid = True, figsize = (12,6));
```

```python
ds1.describe()
```

```python
# % change
ds2 = ds1.pct_change()
ds2.head()
```

```python
ds2.plot(grid = True, figsize = (12,6));
```

```python
ds2.describe()
```

```python
ds2.describe()[1:2]
```

```python
sigma = ds2.cov()
sigma
```

```python
#corrilation
ds2.corr()
```

```python
import seaborn as sns
fig = plt.figure(figsize = (12,5))
sns.heatmap(ds2.corr(), cmap = 'coolwarm')
fig.suptitle('correlation');
```
